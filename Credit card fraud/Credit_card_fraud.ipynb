```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv('/content/creditcard.csv')
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df=df.dropna()
```

```python
df.isnull().sum()
```

```python
df['Class'].value_counts()
```

Highly Imbalanced data

```python
#So we have to seperate the class data
legit=df[df.Class==0]
fraud=df[df.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
legit.Amount.describe()
```

```python
df.groupby('Class').mean()
```

UnderSampling

```python
legit_sample=legit.sample(n=363)
```

Concatinating two Dataframe

```python
new_dataset=pd.concat([legit_sample,fraud],axis=0)
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```

Data seperation

```python
x=new_dataset.drop(columns='Class',axis=1)
y=new_dataset['Class']
print(x.shape)
print(y.shape)
```

Train and test data Split

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,stratify=None,random_state=2)
print(x_train.shape)
print(x_test.shape)
```

Model Training

```python
model=LogisticRegression()
model.fit(x_train,y_train)
```

Model_Evaluation

```python
x_train_prediction=model.predict(x_train)
prediction=accuracy_score(y_train,x_train_prediction)
print(prediction)
```

Model testing

```python
x_test_prediction=model.predict(x_test)
prediction=accuracy_score(y_test,x_test_prediction)
print(prediction)
```
